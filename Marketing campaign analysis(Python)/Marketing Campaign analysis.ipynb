```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv("C:/Users/ACN\Desktop/Marketing campaign analysis(Python)/marketing_campaign.csv")

df.head(10).T
```

```python
df.info()
```

```python
df.dropna(subset=["Income"])
df.head(10).T
```

```python
campaign_cols = ["AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3", "AcceptedCmp4", "AcceptedCmp5"]

df["Total_Campaign_Response"] = df[campaign_cols].sum(axis=1)

acceptance_rate = (df["Total_Campaign_Response"] > 0).mean() * 100

print(acceptance_rate)
```

```python
campaign_performance = df[campaign_cols].sum()
campaign_performance

campaign_performance.plot(
    kind="bar",
    title="Total Acceptances by Campaign"
)
```

```python
response_by_education = df.groupby("Education")["Total_Campaign_Response"].mean()
response_by_education
print(response_by_education)

response_by_education.plot(kind="bar", title="Campaign Response by Education")

```

```python
df.groupby("Marital_Status")["Total_Campaign_Response"].mean().plot(
    kind="bar",
    title="Campaign Response by Marital Status"
)
```
